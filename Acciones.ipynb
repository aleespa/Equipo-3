```python
from __future__ import division
import pandas as pd
import fix_yahoo_finance as yf
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
SP = list(pd.read_csv('Stickers.csv',header=None)[0])
```

```python
pnl = yf.download(SP,start="2017-01-01")
```

```python

df = pnl.to_frame().unstack(level=1)
df = df['Adj Close']
```

```python
df.to_csv('S&P500_Data.csv')
```

```python
df.describe()
```

```python
sd=[]
me=[]
minn=[]
maxx=[]
aux=0
aux1=0
aux2=0
aux3=0
for i in (df.columns):
    aux=np.std(list(df[i]))
    aux1=np.mean(list(df[i]))
    aux2=np.min(list(df[i]))
    aux3=np.max(list(df[i]))
    sd.append(aux)
    me.append(aux1)
    minn.append(aux2)
    maxx.append(aux3)
   
da=pd.read_csv('Sectores.csv')
```

```python
da['Mean']=me
da['DesEst']=sd
da['Min']=minn
da['Max']=maxx
da.head(70)

```

```python
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import MDS
```

```python
var=['DesEst','Mean','Min','Max']
X=da[var].copy()
sc = StandardScaler()
mm = MinMaxScaler()
```

```python
sc.fit(X)
mm.fit(X)
```

```python
Xs = pd.DataFrame(sc.transform(X),columns=X.columns)
Xmm = pd.DataFrame(mm.transform(X),columns=X.columns)
```

```python
pca = PCA(n_components=2)
pca.fit(Xs)
Xp = pd.DataFrame(pca.transform(Xs),columns=['p1','p2'])
```

```python
mds = MDS(n_components=2)
Xm = pd.DataFrame(mds.fit_transform(Xmm),columns=['d1','d2'])
```

```python
sns.lmplot(data=Xp,x='p1',y='p2',fit_reg=False)
```

```python
sns.lmplot(data=Xm,x='d1',y='d2',fit_reg=False)
```

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.cluster import KMeans
from sklearn.mixture import GMM
```

```python
lst_inertia = []
for k in range(2,7):
    cl = KMeans(n_clusters=k)
    cl.fit(Xmm)
    lst_inertia.append(cl.inertia_)
```

```python
plt.plot(range(2,7),lst_inertia,marker='o')
```
