```python
import pandas as pd
from __future__ import division
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
%matplotlib inline
import  pickle

from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.manifold import MDS

from sklearn.cluster import AgglomerativeClustering, KMeans
from sklearn.mixture import GaussianMixture

from sklearn.feature_selection import SelectKBest
```

```python
df = pd.read_csv('S&P500_2017.csv')
df = df.drop(['Date'],axis=1)
df= df.drop(df.select_dtypes(['object']).columns,axis=1)
df=df.fillna(df.mean())
mm1 = MinMaxScaler()
mm1.fit(df)
dfm = pd.DataFrame(mm1.transform(df),columns=df.columns)

dfm.head()
```

```python
plt.plot(dfm.DHI)
```

```python
dfm = dfm.drop(['WELL','BKNG','BHF'],axis=1).copy()
```

```python
dfm.head()
```

```python
def dias(X):
    y = 0
    for i in range(1,len(X)):
        if (X[i-1]< X[i]):
            y = y + 1     
    return y
def ren(X):
    y = []
    for i in range(1,len(X)):
        y.append((X[i]-X[i-1])/X[i-1])   
    return np.mean(y)

d = []
r = []
for col in dfm.columns:
    d.append(dias(dfm[col]))
    r.append(ren(df[col]))
```

```python
X = pd.DataFrame(columns=dfm.columns)
X = X.T
X['Media'] = dfm.mean()
X['Std'] = dfm.std()
X['Dias'] = d
X['r'] = r
```

```python
X= X.fillna(X.mean())
X.head()
```

```python
sns.heatmap(X.corr())
```

```python
X.hist()
```

```python
sc = pickle.load(open('pipe_sc.m','rb'))
mm = pickle.load(open('pipe_mm.m','rb'))
pca = pickle.load(open('pipe_pca.m','rb'))
mds = pickle.load(open('pipe_mds.m','rb'))
```

```python
Xs = pd.DataFrame(sc.transform(X),columns=X.columns)
Xmm = pd.DataFrame(mm.transform(X),columns=X.columns)
```

```python
pca = PCA(n_components=2)
pca.fit(Xs)
Xp = pd.DataFrame(pca.transform(Xs),columns=['p1','p2'])
```

```python
mds = MDS(n_components=2)
Xm = pd.DataFrame(mds.fit_transform(Xmm),columns=['d1','d2'])
```

```python
sns.lmplot(data=Xp,x='p1',y='p2',fit_reg=False)
```

```python
sns.lmplot(data=Xm,x='d1',y='d2',fit_reg=False)
```

```python
lst_inertia = []
for k in range(2,11):
    cl = KMeans(n_clusters=k)
    cl.fit(Xmm)
    lst_inertia.append(cl.inertia_)
```

```python
plt.plot(range(2,11),lst_inertia,marker='o')
```

```python
cl = pickle.load(open('cluster.m','rb'))
```

```python
Xp['cl']=Xmm['cl'] = Xm['cl']=Xp['cl'] =X['cl']= cl.predict(Xmm)
```

```python
sk = SelectKBest(k=4)
sk.fit(Xmm[['Media','Std','Dias','r']],Xmm['cl'])
```

```python
var_best= [x for x,y in zip(['Media','Std','Dias','r'],sk.get_support()) if y]
var_best
```

```python
sns.heatmap(Xmm[['cl']+var_best].groupby('cl').mean(),annot=True,cmap='RdYlGn')
```
